## PCA 

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
# Datos
X, y = datasets.load_breast_cancer(return_X_y=True)
```

```python
X.shape
```

```python
# Centrar
xNew = StandardScaler().fit_transform(X)
```

```python
# PCA

# Calcular matriz de covarianza
matriz_cov = np.cov(xNew.T) 

# Calcular vectores y valores propios
eig_val, eig_vecs = np.linalg.eig(matriz_cov)
```

```python
# Primer componente 
X_proy    = xNew.dot(eig_vecs.T[0])
eje_dummy = np.zeros(y.shape[0])


ax=plt.gca()
ax.scatter(X_proy,eje_dummy, c=y, cmap='Paired')
plt.xlabel('pc1')
plt.ylabel('dummy value')
plt.show()
```

```python
###### 2 componentes

X_proy1   = xNew.dot(eig_vecs.T[0])
X_proy2   = xNew.dot(eig_vecs.T[1])

ax=plt.gca()
ax.scatter(X_proy1,X_proy2,c=y, cmap='Paired')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.show()
```

```python
###### 3 componentes

X_proy1   = xNew.dot(eig_vecs.T[0])
X_proy2   = xNew.dot(eig_vecs.T[1])
X_proy3   = xNew.dot(eig_vecs.T[2])

ax = plt.axes(projection='3d')
ax.set_xlabel('pc1')
ax.set_ylabel('pc2')
ax.set_zlabel('pc3')
ax.scatter3D(X_proy1, X_proy2, X_proy3, c=y, cmap='Paired');
plt.show()
```
