##  EM - Ejemplo 1

```python
# Librerias
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
import numpy as np
```

```python
# Datos
np.random.seed(1)
 
muestras = 200
X,Y = make_blobs(centers=3,cluster_std=1.5,n_samples=muestras)
plt.scatter(X[:, 0], X[:, 1], c='green')
plt.show()
```

```python
# Calculo de Clusters  
M = 4
GMM = GaussianMixture(n_components=M).fit(X)

# Predicción
YH = GMM.predict(X)

# Mostrar clusters calculados
ax=plt.gca()
ax.scatter(GMM.means_[:,0], GMM.means_[:,1], c=range(GMM.means_.shape[0]), ec='k', s=200)
ax.scatter(X[:, 0], X[:, 1], c=YH, ec='k')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title("Prediccion con M = " + str(M))
plt.show()
```

```python
# Mostrar clusters "originales"
ax=plt.gca()
ax.scatter(X[:, 0], X[:, 1], c=Y, ec='k')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title("Real")
plt.show()
```
