# Regresión lineal - Implementación

## Formula 

$ y = ax + b$

```python
# Librerias

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos dummy

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
m = x.size

plt.plot(x,y,'o')
plt.show()
```

```python
# Inicializar

a        = np.random.rand()
b        = np.random.rand()
alpha    = 0.1  # Tasa de aprendizaje 
maxItera = 1000
i        = 0     

loga = np.array([a])
logb = np.array([b])
```

```python
# Ajuste de parámetros - Aprendizaje DG
while i < maxItera:
    # Evaluar - y = ax + b
    yF = a * x + b
    Error = yF - y
    
    # Actualizar
    # a <- a - alpha * (1 / m) * sum(Error * x)
    nuevoA = a - alpha * (1/m) * sum(Error * x)
    
    # b <- b - alpha * (1 / m) * sum(Error)
    nuevoB = b - alpha * (1/m) * sum(Error)
    
    a = nuevoA
    b = nuevoB


    # Log
    loga = np.append(loga,a)
    logb = np.append(logb,b)

    i += 1
    
```

```python
plt.plot(x,y,'o')
plt.plot(x,a*x + b)
plt.show()
```

```python
plt.plot(loga, label='a')
plt.plot(logb, label='b')
plt.legend()
plt.show()
```
