# Regresión lineal - Regularización Lasso

## Formula 

$ y = \theta_0 + \theta_1 x + \theta_2 x^2$

```python
# Librerias

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos
from sklearn import datasets
xT, y = datasets.load_diabetes(return_X_y=True)
x = xT[:,6]
m = x.size
plt.plot(x,y,'o')
```

```python
### Aprendizaje 
from sklearn import linear_model

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)


X = x.reshape(-1,1)
Y = y.reshape(-1,1)

Xpoly = poly.fit_transform(X)

modeloRL = linear_model.LinearRegression()
modeloRL.fit(Xpoly,Y)
print(" Modelo 1 Sklearn ")
print("theta0: ", modeloRL.intercept_)
print("theta1-2: ", modeloRL.coef_)
```

```python
# Predicción scikit-learn
plt.plot(X,Y,'o')
plt.plot(X,modeloRL.intercept_ + modeloRL.coef_[0,0]*X + modeloRL.coef_[0,1]*X**2,'o') 
plt.show()
```

```python
# Aprendizaje con regularización Lasso
from sklearn import linear_model
modelo = linear_model.Lasso(alpha=1.0)

modelo.fit(Xpoly, Y)
```

```python
print(" Modelo Sklearn ")
print("theta0: ", modelo.intercept_)
print("theta1: ", modelo.coef_)
```

```python
# Predicción scikit-learn
plt.plot(X,Y,'o')
plt.plot(X,modelo.intercept_ + modelo.coef_[0]*X + modelo.coef_[1]*X**2,'o') 
plt.show()
```
