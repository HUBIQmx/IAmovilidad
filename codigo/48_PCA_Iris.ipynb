## PCA - Iris dataset

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
# Datos
X, y = datasets.load_iris(return_X_y=True)
```

```python
X.shape
```

```python
# Centrar
xNew = StandardScaler().fit_transform(X)
```

```python
# PCA

# Calcular matriz de covarianza
matriz_cov = np.cov(xNew.T) 

# Calcular vectores y valores propios
eig_val, eig_vecs = np.linalg.eig(matriz_cov)
```

```python
###### 2 componentes

X_proy1   = xNew.dot(eig_vecs.T[0])
X_proy2   = xNew.dot(eig_vecs.T[1])

ax=plt.gca()
ax.scatter(X_proy1,X_proy2,c=y, cmap='Paired')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit(X).transform(X)
```

```python
ax=plt.gca()
ax.scatter(X_pca[:,0],X_pca[:,1],c=y, cmap='Paired')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.show()
```
