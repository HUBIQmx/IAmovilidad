## DBSCAN - Ejemplo 2

```python
# Librerias
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Datos
np.random.seed(1)
muestras = 200
X,Y = make_moons(n_samples=muestras,noise=0.1)
X = StandardScaler().fit_transform(X)

plt.scatter(X[:, 0], X[:, 1], c='green',s=50)
plt.show()
```

```python
# DBSCAN
modelo = DBSCAN(eps=0.4, min_samples=10).fit(X)
y2     = modelo.labels_

plt.scatter(X[:, 0], X[:, 1], c=y2, ec='k',s=50)
plt.show()
```

```python
# Mostrar clusters "originales"
ax=plt.gca()
ax.scatter(X[:, 0], X[:, 1], c=Y, ec='k',s=50)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title("Real")
plt.show()
```
