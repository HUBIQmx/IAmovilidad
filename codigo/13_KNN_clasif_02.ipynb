# K-vecinos más cercanos - Clasificación

## Formula 

$ Pr (Y = j | X = x_0) = \frac{1}{k} \sum_{x^i \in N_{k}(x_0)} I(y^i = j) $

```python
# Librerias
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
import math as math
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
```

```python
def normaliza(X):
    media = X.mean()
    std   = X.std()
    newX = (X - media) / std
    return newX
```

```python
# Datos  
np.random.seed(2)
m = 100 

# Entrenamiento y prueba
X,Y = make_blobs(centers=2,cluster_std=1.0,random_state=3,n_samples=m)
X_t1, X_t2, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Normalizamos datos
X_train = normaliza(X_t1)
X_test  = normaliza(X_t2)

colors=['red' if i==0 else 'blue' for i in y_train]
ax=plt.gca()
ax.scatter(X_train[:, 0], X_train[:, 1], c=colors)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title("Datos de Entrenamiento")
plt.show()

colors=['red' if i==0 else 'blue' for i in y_test]
ax=plt.gca()
ax.scatter(X_test[:, 0], X_test[:, 1], c=colors)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title("Datos de Prueba")
plt.show()
```

```python
# Modelo
knn = KNeighborsClassifier(n_neighbors=11) # k
knn.fit(X_train, y_train)
```

```python
print(knn.predict(X_test))
```

```python
print(knn.predict_proba(X_test))
```

```python
colors=['red' if i==0 else 'blue' for i in y_test]
ax=plt.gca()
ax.scatter(X_test[:, 0], X_test[:, 1], c=colors)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title("Datos de Prueba")
plt.show()

colors=['red' if i==0 else 'blue' for i in knn.predict(X_test)]
ax=plt.gca()
ax.scatter(X_test[:, 0], X_test[:, 1], c=colors)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title("Datos de Prueba")
plt.show()
```
