## DBSCAN - Ejemplo 1 

```python
# Librerias
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Datos
np.random.seed(1) 
muestras = 200
X,Y = make_blobs(centers=3,cluster_std=1.5,n_samples=muestras)
X = StandardScaler().fit_transform(X)

plt.scatter(X[:, 0], X[:, 1], c='green',s=50)
plt.show()
```

```python
# DBSCAN
modelo = DBSCAN(eps=0.7, min_samples=5).fit(X)
y2     = modelo.labels_

plt.scatter(X[:, 0], X[:, 1], c=y2, ec='k',s=50)
plt.show()
```
