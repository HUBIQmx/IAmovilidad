# K-vecinos más cercanos - Regresión

## Formula 

$ y = \frac{1}{k} \sum_{x^i \in N_{k}(x_0)} y^i $

```python
# Librerias
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
def normaliza(X):
    media = X.mean()
    std   = X.std()
    newX = (X - media) / std
    return newX
```

```python
# Datos  
from sklearn import datasets
xT, y = datasets.load_diabetes(return_X_y=True)
x = xT[:,6]

# Entrenamiento y prueba
X_t1, X_t2, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Normalizamos datos
X_train = normaliza(X_t1)
X_test  = normaliza(X_t2)

X_train = X_train.reshape(-1,1)
X_test  = X_test.reshape(-1,1)
y_train = y_train.reshape(-1,1)
y_test  = y_test.reshape(-1,1)

plt.plot(X_train,y_train,'o')
plt.show()

plt.plot(X_test,y_test,'o')
plt.show()
```

```python
# Modelo
knn = KNeighborsRegressor(n_neighbors=3) # k
knn.fit(X_train, y_train)
```

```python
plt.plot(X_test,y_test,'o')
plt.plot(X_test,knn.predict(X_test),'+')
plt.show()
```
