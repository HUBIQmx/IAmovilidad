# K-vecinos más cercanos - Regresión

## Formula 

$ y = \frac{1}{k} \sum_{x^i \in N_{k}(x_0)} y^i $

```python
# Librerias
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
```

```python
# Datos  
from sklearn import datasets
xT, y = datasets.load_diabetes(return_X_y=True)
x = xT[:,6]

X=(x-min(x))/(max(x)-min(x)) 
Y=(y-min(y))/(max(y)-min(y))

X = X.reshape(-1,1)
Y = Y.reshape(-1,1)

plt.plot(X,Y,'o')
```

```python
# Modelo
knn = KNeighborsRegressor(n_neighbors=5) # k
knn.fit(X, Y)
```

```python
print(knn.predict([[0.25]]))
```
